```python
import numpy as np
df['TARGET CLASS'] = np.where(df['TARGET CLASS'] > 0.5, 1, 0)
```

```python
df
```

```python
features = df.drop(columns=['TARGET CLASS'])
```

```python
target = df['TARGET CLASS']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

features = scaler.fit_transform(features)
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(features,target,test_size=0.2,shuffle=True,random_state=21)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train,y_train)

y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)

print(accuracy)


```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(cm,annot=True)
plt.show()
```
